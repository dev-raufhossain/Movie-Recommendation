```python
new_df['actor_1_name'] = new_df['actor_1_name'].astype(str).apply(lambda x: x.split())
new_df.head()
```

```python
new_df['actor_1_name'] = new_df['actor_1_name'].apply(lambda x:["".join(x)])
new_df['actor_1_name']
```

```python
new_df[['director_name','actor_2_name','actor_3_name']] = new_df[['director_name','actor_2_name','actor_3_name']].astype(str).apply(lambda x:[i.split() for i in x])
new_df[['director_name','actor_2_name','actor_3_name']] = new_df[['director_name','actor_2_name','actor_3_name']].apply(lambda x:[["".join(i)] for i in x])
new_df.sample(5)
```

```python
new_df['movie_title'] = new_df['movie_title'].astype(str).apply(lambda x: x.split())
new_df['movie_title']
```

```python
new_df[['genres','plot_keywords']] = new_df[['genres','plot_keywords']].astype(str).apply(lambda x:[i.split("|") for i in x])
new_df.head()
```

```python
new_df['imdb_score'] = new_df['imdb_score'].astype(str).apply(lambda x: x.split())
new_df.imdb_score
```

```python
new_df.sample(5)
```

```python
new_df['tag'] = new_df['movie_title']+new_df['director_name']+new_df['actor_1_name']+new_df['actor_2_name']+new_df['actor_3_name']+new_df['genres']+new_df['plot_keywords']+new_df['imdb_score']
new_df.head()
```

```python
import re
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x:[re.sub(":|-|/|=|,|",'',i) for i in x])
new_df['tag'][4]
```

```python
df = new_df.loc[:,('movie_title', 'tag', 'movie_imdb_link')]
df.head()
```

```python
df['movie_title'] = df['movie_title'].apply(lambda x: " ".join(x))
df['movie_title']
```

```python
df['tag'] = df['tag'].apply(lambda x: " ".join(x))
df['tag']
```

```python
df.head()
```

```python
df['tag'] = df['tag'].apply(lambda x: x.lower())
df['tag']
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def steming(text):
    word = []
    for i in text.split():
        s = ps.stem(i)
        word.append(s)
    return " ".join(word)
```

```python
df['tag'] = df['tag'].apply(lambda x: steming(x))
df['tag'][0]
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(df['tag']).toarray()
vectors
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity[2]
```

```python
list(enumerate(similarity[1]))
```

```python
sorted(list(enumerate(similarity[1])),reverse=True,key=(lambda x: x[1]))
```

```python
def recommend(movie):
    movie_index = df[df['movie_title'] == movie].index[0]
    movies_list = sorted(list(enumerate(similarity[movie_index])),reverse=True,key=(lambda x: x[1]))[1:6]
   # print(movies_list)
    for i in movies_list:
        movies = df.iloc[i[0]]
        print(movies)
    return
```

```python
recommend('Avatar')
```
